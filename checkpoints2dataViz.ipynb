```python

import seaborn as sns
import matplotlib.pyplot as plt

def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }
        )
plot_correlation_map(data)
```

The resulting graph describes the correlations between the different features.<br>
The resulting values varies between 0 and 1.<br>
the coefficient of correlation between a column and itself is equal to one. <br>
There are columns that are positively correlated. Others are negatively correlated.
